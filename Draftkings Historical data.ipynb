```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
r = urllib.urlopen('http://rotoguru1.com/cgi-bin/nba-dhd-2018.pl').read()
soup = BeautifulSoup(r)
```

```python
raw = soup.find('pre').text
raw = raw.split('\n')
data = []
print "records num: " + str(len(raw))
for line in raw:
    data.insert(0, line.split(':')[:-1])
print data[0: 10]
df = pd.DataFrame(data[1:], columns=data[0])
print df.head()
print df.describe()
```

```python
print '======= Louis Willams DK salary vs DK Points ============='
lw_dk = df.loc[df['First  Last'] == 'Louis Williams'][['Date', 'DK Sal', 'DKP']]
lw_dk[lw_dk['DK Sal'] != ''][['DK Sal']].astype(float).plot()
lw_dk[lw_dk['DKP'] != ''][['DKP']].astype(float).plot()
```

```python
print '======= Louis Willams DK salary vs FD salary vs YH salary ============='
lw_dk = df.loc[df['First  Last'] == 'Louis Williams'][['Date', 'DK Sal', 'FD Sal', 'YH Sal']]
lw_dk[lw_dk['DK Sal'] != ''][['DK Sal']].astype(float).plot()
lw_dk[lw_dk['FD Sal'] != ''][['FD Sal']].astype(float).plot()
lw_dk[lw_dk['YH Sal'] != ''][['YH Sal']].astype(float).plot()
```
